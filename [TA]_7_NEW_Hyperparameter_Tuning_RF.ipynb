```
#Import Library
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, confusion_matrix
import time
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

#Skenario 3

```
#DATASET
train_full_uni = pd.read_csv("drive/My Drive/TA/TF-IDF/Translated/data_train_trans_full_uni.csv")
test_full_uni = pd.read_csv("drive/My Drive/TA/TF-IDF/Translated/data_test_trans_full_uni.csv")
```

```
#Tagging

X_train = train_full_uni.drop(['harga_','pengemasan_','produk_','aroma_'],axis=1)
y_train = train_full_uni[['harga_','pengemasan_','produk_','aroma_']]

X_test = test_full_uni.drop(['harga_','pengemasan_','produk_','aroma_'],axis=1)
y_test = test_full_uni[['harga_','pengemasan_','produk_','aroma_']]
```

```
def rfc(X_train, X_test, y_train, y_test, n_est, max_Features, depth, criteria):
    #%%time
    classifier_MO = MultiOutputClassifier(RandomForestClassifier(n_estimators=n_est, max_depth=depth, max_features=max_Features, criterion=criteria))
    classifier_MO.fit(X_train, y_train)
    y_pred = classifier_MO.predict(X_test)
    #pred = rf.predict(X_train) # predict data train
    return X_train, X_test, y_train, y_test, y_pred
```

```
def accuracy(y_test, y_pred):
    scoring = []
    harga = []
    pengemasan = []
    produk = []
    aroma = []
    acc_all = []
        
    harga_acc = accuracy_score(y_test['harga_'],y_pred[:,0])
    kemas_acc = accuracy_score(y_test['pengemasan_'],y_pred[:,1])
    produk_acc = accuracy_score(y_test['produk_'],y_pred[:,2])
    aroma_acc = accuracy_score(y_test['aroma_'],y_pred[:,3])

    acc = (harga_acc + kemas_acc + produk_acc + aroma_acc)/4
    harga.append(harga_acc)
    pengemasan.append(kemas_acc)
    produk.append(produk_acc)
    aroma.append(aroma_acc)
    scoring.append(acc)

    acc_all.append(np.mean(harga))
    acc_all.append(np.mean(pengemasan))
    acc_all.append(np.mean(produk))
    acc_all.append(np.mean(aroma))
    acc_all.append(np.mean(scoring))

    return acc_all
```

```
def precision(y_test, y_pred):
    scoring = []
    harga = []
    pengemasan = []
    produk = []
    aroma = []
    pre_all = []
        
    harga_pre = precision_score(y_test['harga_'],y_pred[:,0], average='macro')
    kemas_pre = precision_score(y_test['pengemasan_'],y_pred[:,1], average='macro')
    produk_pre = precision_score(y_test['produk_'],y_pred[:,2], average='macro')
    aroma_pre = precision_score(y_test['aroma_'],y_pred[:,3], average='macro')

    pre = (harga_pre + kemas_pre + produk_pre + aroma_pre)/4
    harga.append(harga_pre)
    pengemasan.append(kemas_pre)
    produk.append(produk_pre)
    aroma.append(aroma_pre)
    scoring.append(pre)

    pre_all.append(np.mean(harga))
    pre_all.append(np.mean(pengemasan))
    pre_all.append(np.mean(produk))
    pre_all.append(np.mean(aroma))
    pre_all.append(np.mean(scoring))

    return pre_all
```

```
def recall(y_test, y_pred):
    scoring = []
    harga = []
    pengemasan = []
    produk = []
    aroma = []
    rec_all = []
        
    harga_re = recall_score(y_test['harga_'],y_pred[:,0], average='macro')
    kemas_re = recall_score(y_test['pengemasan_'],y_pred[:,1], average='macro')
    produk_re = recall_score(y_test['produk_'],y_pred[:,2], average='macro')
    aroma_re = recall_score(y_test['aroma_'],y_pred[:,3], average='macro')

    re = (harga_re + kemas_re + produk_re + aroma_re)/4
    harga.append(harga_re)
    pengemasan.append(kemas_re)
    produk.append(produk_re)
    aroma.append(aroma_re)
    scoring.append(re)

    rec_all.append(np.mean(harga))
    rec_all.append(np.mean(pengemasan))
    rec_all.append(np.mean(produk))
    rec_all.append(np.mean(aroma))
    rec_all.append(np.mean(scoring))

    return rec_all
```

```
def recall(y_test, y_pred):
    scoring = []
    harga = []
    pengemasan = []
    produk = []
    aroma = []
    rec_all = []
        
    harga_re = recall_score(y_test['harga_'],y_pred[:,0], average='macro')
    kemas_re = recall_score(y_test['pengemasan_'],y_pred[:,1], average='macro')
    produk_re = recall_score(y_test['produk_'],y_pred[:,2], average='macro')
    aroma_re = recall_score(y_test['aroma_'],y_pred[:,3], average='macro')

    re = (harga_re + kemas_re + produk_re + aroma_re)/4
    harga.append(harga_re)
    pengemasan.append(kemas_re)
    produk.append(produk_re)
    aroma.append(aroma_re)
    scoring.append(re)

    rec_all.append(np.mean(harga))
    rec_all.append(np.mean(pengemasan))
    rec_all.append(np.mean(produk))
    rec_all.append(np.mean(aroma))
    rec_all.append(np.mean(scoring))

    return rec_all
```

```
def f1score(y_test, y_pred):
    scoring = []
    harga = []
    pengemasan = []
    produk = []
    aroma = []
    f1_arr = []
        
    harga_f1 = f1_score(y_test['harga_'],y_pred[:,0], average='macro')
    kemas_f1 = f1_score(y_test['pengemasan_'],y_pred[:,1], average='macro')
    produk_f1 = f1_score(y_test['produk_'],y_pred[:,2], average='macro')
    aroma_f1 = f1_score(y_test['aroma_'],y_pred[:,3], average='macro')

    avg_f1 = (harga_f1 + kemas_f1 + produk_f1 + aroma_f1)/4 #
    harga.append(harga_f1)
    pengemasan.append(kemas_f1)
    produk.append(produk_f1)
    aroma.append(aroma_f1)
    scoring.append(avg_f1)

    f1_arr.append(np.mean(harga))
    f1_arr.append(np.mean(pengemasan))
    f1_arr.append(np.mean(produk))
    f1_arr.append(np.mean(aroma))
    f1_arr.append(np.mean(scoring))

    return f1_arr
```
